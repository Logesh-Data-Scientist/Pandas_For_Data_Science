```python
import pandas as pd

data = {
    "ID": [101, 102, 103, 104],
    "Name": ["Logesh", "John", "Alice", "Bob"],
    "Salary": [30000, 50000, 45000, 60000]
}

df = pd.DataFrame(data)
print(df)
```

## df.set_index()

```python
df_indexed = df.set_index("ID")
print(df_indexed)
```

## df.reset_index()

```python
df_reset = df_indexed.reset_index()
print(df_reset)
```

## df.reindex()

```python
# Change index order and add new index
new_index = [104, 103, 102, 101, 105]

df_reindexed = df_indexed.reindex(new_index)
print(df_reindexed)
```
