```python
import pandas as pd

data = {
    "Name": ["Logesh", "John", "Alice", "Bob", "Charlie"],
    "Gender": ["Male", "Male", "Female", "Male", "Female"],
    "Marks": [45, 67, 89, 55, 72]
}

df = pd.DataFrame(data)
df
```

## pd.Categorical()

```python
df["Gender"] = pd.Categorical(df["Gender"], categories=["Male", "Female"])
df.dtypes
```

## pd.get_dummies()

```python
pd.get_dummies(df["Gender"], drop_first=True)
```

## pd.cut()

```python
bins = [0, 50, 70, 100]
labels = ["Low", "Medium", "High"]

df["Marks_Category"] = pd.cut(df["Marks"], bins=bins, labels=labels)
df
```

## pd.qcut()

```python
df["Marks_Quantile"] = pd.qcut(df["Marks"], q=3)
df
```

## df.cat.*

```python
# Category codes
print(df["Gender"].cat.codes)

# Categories
print(df["Gender"].cat.categories)

# Add new category
df["Gender"] = df["Gender"].cat.add_categories(['other'])
print(df["Gender"].cat.categories)
```
