```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Logesh", "John", "Alice", "Bob", "John", "Alice"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT"],
    "Salary": [30000, 50000, 45000, 60000, 52000, 48000],
    "Date": pd.date_range(start="2024-01-01", periods=6, freq="D")
}

df = pd.DataFrame(data)
```

```python
df.set_index("Date", inplace=True)
print(df)
```

## df.groupby()

```python
print(df.groupby("Department")["Salary"].mean())
```

## .agg()

```python
print(df.groupby("Department")["Salary"].agg(["mean", "sum", "max"]))
```

## df.pivot_table()

```python
print(pd.pivot_table(df, values="Salary", index="Department", aggfunc="mean"))
```

## df.resample()

```python
print(df.resample("2D")["Salary"].sum())
```

## df.rolling()

```python
print(df["Salary"].rolling(window=2).mean())
```
