```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\Hey!\OneDrive\Desktop\Pandas-for-Data-Science\dataset\countries of the world.csv")
```

```python
df2 = pd.read_csv(r"C:\Users\Hey!\OneDrive\Desktop\Pandas-for-Data-Science\dataset\world_population.csv")
```

## df.merge()

```python
# Inner Join
print(pd.merge(df1, df2, on="Country", how="inner"))
```

```python
# Left Join
print(pd.merge(df1, df2, on="Country", how="left"))
```

```python
# Right Join
print(pd.merge(df1, df2, on="Country", how="right"))
```

```python
# Outer Join
print(pd.merge(df1, df2, on="Country", how="outer"))
```

## pd.concat()

```python
# Row-wise concat
print(pd.concat([df1, df2], axis=0))
```

```python
# Column-wise concat
print(pd.concat([df1, df2], axis=1))
```

## df.join()

```python
# Set index for join
df1_indexed = df1.set_index("Country")
df2_indexed = df2.set_index("Country")

print(df1_indexed.join(df2_indexed, how="left"))
```
