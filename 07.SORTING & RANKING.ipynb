```python
import pandas as pd
```

```python
data = {
    "Name": ["Bruce", "Harvy", "joker", "bane", "cordern"],
    "Age": [21, 25, 23, 28, 26],
    "Salary": [30000, 50000, 45000, 60000, 52000]
}

df = pd.DataFrame(data)
df
```

## df.sort_values()

```python
# Sort by single column
df.sort_values(by="Salary")  # defualt order 'ascending'
```

```python
# Sort descending
df.sort_values(by="Salary", ascending=False)
```

```python
# Sort by multiple columns
df.sort_values(by=["Age", "Salary"])
```

## df.sort_index()

```python
# Shuffle index first
df_shuffled = df.sample(frac=1)
```

```python
# Sort by index
df_shuffled.sort_index()
```

## df.rank()

```python
# Rank salaries
df["Salary"].rank()
```

```python
# Rank with highest as rank 1
df["Salary"].rank(ascending=False)
```

## df.nlargest()

```python
# Top 3 highest salaries
df.nlargest(3, "Salary")
```

## df.nsmallest()

```python
# Bottom 2 lowest salaries
df.nsmallest(2, "Salary")
```
